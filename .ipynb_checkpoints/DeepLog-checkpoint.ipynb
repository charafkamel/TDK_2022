```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
```

### Adatok Betöltése

```python
rawdata = pd.read_csv('HDFS.csv')
```

```python
raw_templates = pd.read_csv('HDFS_templates.csv')
raw_anomalies = pd.read_csv('anomaly_label.csv')
every_EventId = {}
for i in range(0, len(raw_templates)):
    every_EventId[raw_templates.iloc[i][0]] = i+1
every_EventId
```

### Szükséges oszlopok megtartása

```python
filtered_data = rawdata[['LineId', 'EventId', 'ParameterList']] #Kivesszük az értelmetlen infókat
```

### Blokkazonosítók meghatározása

```python
block_Event_dict = {} ##
listank = set()
for i in range(0,len(filtered_data)):
    raw_ParamterList = filtered_data.iloc[i][2] # Kiválasztjuk a Paraméterlistát
    first_occurrence = raw_ParamterList.find("blk_") + 4 # Megtaláljuk a blk_-t és hozzáadtunk 4-et, mert lk_ és utána kezdődjön
    block_azon_string = raw_ParamterList[first_occurrence: first_occurrence + 10]# megkapjuk a stringet 19 (blockazon_hossza)
    block_azon = int(block_azon_string) # átállunk integerekre
    
    if block_azon not in listank: # Ha nincs benne eddig már ez a blockazonosító
        listank.add(block_azon) # Adjuk hozzá a halmazhoz
        block_Event_dict[block_azon] = [every_EventId.get(filtered_data.iloc[i][1])] # Hozzuk létre a dict-ben a kulcs-értéktömb párt
        
    else: # Ha benne van,
        block_Event_dict.get(block_azon).append(every_EventId.get(filtered_data.iloc[i][1])) # akkor adjuk hozzá a megfelelő értéktömbhöz az elemet
block_Event_dict
```

### Training és Teszt adatok létrehozása

```python
lista = block_Event_dict.keys()
kulcsok_hossza = 0
if len(lista) % 2 == 1: 
    kulcsok_hossza = (len(lista)-1)/2
else:
    kulcsok_hossza = len(lista)/2

data_training = dict(list(block_Event_dict.items())[:int(kulcsok_hossza)])
data_test =   dict(list(block_Event_dict.items())[int(kulcsok_hossza): len(lista)])
```

### Mátrixok létrehozása

```python
##scaler = MinMaxScaler()
##training_data = scaler.fit_transform(training_data)
X_train = []
y_train = []

list_temp = [] 

maxlen = 0
MaxWindow = 50

for i in range(0, len(data_training.keys())):
    X_train.append([])
    first_value = list(data_training.values())[i] #Adott kulcsokhoz tartozó tömb
    if len(first_value) < MaxWindow: ## Ha kisebb mint a MaxWindow
        list_of_zeros = [0] * (MaxWindow - len(first_value)) # Zérusokat tartalmazó
        X_to_append = first_value + list_of_zeros
        X_train.append(X_to_append)
        
    Y_to_append = X_to_append[1:] 
    Y_to_append.append(len(every_EventId.keys()) + 1) # END jel a végén
    y_train.append(Y_to_append)
X_train, y_train = np.array(X_train), np.array(y_train)
print(X_train[0:5])
print("\n ---- \n")
print(y_train[0:5])
```

### LSTM

```python
regressior = Sequential()

regressior.add(LSTM(units = 40, activation = 'relu', return_sequences = True, input_shape = (X_train.shape[1],1,))) ## mert 60 sor és minden sor 5 paraméteres
regressior.add(Dropout(0.2))

regressior.add(LSTM(units = 55, activation = 'relu', return_sequences = True)) ## mert 60 sor és minden sor 5 paraméteres
regressior.add(Dropout(0.2))

regressior.add(LSTM(units = 80, activation = 'relu')) ## mert 60 sor és minden sor 5 paraméteres
regressior.add(Dropout(0.2))
               
regressior.add(Dense(units = 1))
```

```python
regressior.summary() ## Összefoglalja a hálónkat
```

```python
regressior.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
regressior.fit(X_train, y_train, epochs = 5, batch_size = 128)
```
